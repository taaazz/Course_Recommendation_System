```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_compound_score(text):
    if isinstance(text, float):  
        text = str(text)  
    vs = analyzer.polarity_scores(text)
    return vs['compound']

df['reviews'] = df['reviews'].fillna('')
df['compound'] = df['reviews'].apply(get_compound_score)

# Function to classify sentiment into Positive, Neutral, Negative
def classify_sentiment(compound):
    if compound > 0.001:
        return 'Positive', 1
    elif -0.001 < compound < 0.001:
        return 'Neutral', 0
    else:
        return 'Negative', -1

# Apply the overall sentiment classification
df[['Sentiment', 'polarity']] = df['compound'].apply(lambda x: pd.Series(classify_sentiment(x)))

# Function to classify sentiment into an integer value from 1 to 5
def refined_sentiment_classification(vader_compound_score):
    if vader_compound_score >= -1 and vader_compound_score < -0.5:
        return 1
    elif vader_compound_score >= -0.5 and vader_compound_score < -0.05:
        return 2
    elif vader_compound_score >= -0.05 and vader_compound_score <= 0.05:
        return 3
    elif vader_compound_score > 0.05 and vader_compound_score < 0.5:
        return 4
    else:
        return 5

# Apply the refined sentiment classification
df['refined_sentiment'] = df['compound'].apply(refined_sentiment_classification)


```

```python
# Initialize the stopword list and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    if isinstance(text, pd.Series):
        text = text.astype(str)
        text = text.apply(lambda x: re.sub(r'_+', ' ', x))
        text = text.apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
        text = text.str.lower()
        text = text.apply(lambda x: ''.join([c for c in x if c not in punctuation]))
        text = text.apply(lambda x: re.sub(r'\d+', '', x))
        text = text.str.strip()
        text = text.apply(lambda x: re.sub(' +', ' ', x))
        
        # Remove stopwords and lemmatize
        text = text.apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split() if word not in stop_words]))
        
    else:
        text = re.sub(r'_+', ' ', str(text))
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        text = text.lower()
        text = ''.join([c for c in text if c not in punctuation])
        text = re.sub(r'\d+', '', text)
        text = re.sub(' +', ' ', text).strip()
        
        # Remove stopwords and lemmatize
        text = ' '.join([lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words])
        
    return text

# Clean the Sequence column
df['Sequence'] = clean_text(df['Sequence'])
df
```

```python
# Fit the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_encoded = tfidf_vectorizer.fit_transform(df['Sequence'])
```

```python
EMBEDDING_SIZE = 50
# IMPORTANT : pleae read the documentation for register the custom class model to avoid error
# documentation for custom class : https://colab.research.google.com/github/keras-team/keras-io/blob/master/guides/ipynb/serialization_and_saving.ipynb#config_methods
# clear the previoues custom objects
keras.saving.get_custom_objects().clear()
# register the custom layer 
@keras.saving.register_keras_serializable(package="RecomendationLayer")
class RecommenderNet(tf.keras.Model):
    def __init__(self, num_users, num_courses, embedding_size, **kwargs):
        super().__init__(**kwargs)
        self.num_users = num_users
        self.num_courses = num_courses
        self.embedding_size = embedding_size
        self.dense1 = tf.keras.layers.Dense(128, activation="relu")
        self.dense2 = tf.keras.layers.Dense(64, activation="relu")
        self.dense3 = tf.keras.layers.Dense(1, activation="sigmoid")
        self.user_embedding = tf.keras.layers.Embedding(
            num_users,
            embedding_size,
            embeddings_initializer="he_normal",
            embeddings_regularizer=tf.keras.regularizers.l2(1e-6),
        )
        self.user_bias = tf.keras.layers.Embedding(num_users, 1)
        self.food_embedding = tf.keras.layers.Embedding(
            num_courses,
            embedding_size,
            embeddings_initializer="he_normal",
            embeddings_regularizer= tf.keras.regularizers.l2(1e-6),
        )
        self.food_bias = tf.keras.layers.Embedding(num_courses, 1)
    
    def call(self, inputs):
        user_vector = self.user_embedding(inputs[:, 0])
        user_bias = self.user_bias(inputs[:, 0])
        food_vector = self.food_embedding(inputs[:, 1])
        food_bias = self.food_bias(inputs[:, 1])
        dot_user_movie = tf.tensordot(user_vector, food_vector, 2)
        x = dot_user_movie + user_bias + food_bias
        dense_layer1 = self.dense1(x)
        dense_layer2 = self.dense2(dense_layer1)
        return self.dense3(dense_layer2)
    
    def get_config(self):
        config = super().get_config()
        config.update({
            "num_users": self.num_users,
            "num_courses": self.num_courses,
            "embedding_size": self.embedding_size,
        })
        return config
    
    @classmethod
    def from_config(cls, config):
        return  cls( **config)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import tensorflow as tf

def get_train_test_data(df: pd.DataFrame): 
    # Ensure 'refined_sentiment' is numeric, setting errors='coerce' to handle non-numeric entries as NaN
    df['refined_sentiment'] = pd.to_numeric(df['refined_sentiment'], errors='coerce')

    # Drop rows with NaN values in 'refined_sentiment'
    df = df.dropna(subset=['refined_sentiment'])

    x = df[['user_id', 'course_id']].values
    train_len = int(0.98 * df.shape[0])
    
    # Find the max rating after converting to numeric
    max_rating = df["refined_sentiment"].max()
    
    # Normalize 'refined_sentiment' by dividing by max_rating
    y = df['refined_sentiment'].apply(lambda x: x / max_rating).values

    x_train, x_test, y_train, y_test = (
        x[:train_len],
        x[train_len:],
        y[:train_len],
        y[train_len:],
    )
    return x_train, x_test, y_train, y_test 

def get_course_user_len(df: pd.DataFrame): 
    len_user = len(df['user_id'].unique().tolist())
    len_courses = len(df['course_id'].unique().tolist())
    return len_user, len_courses

def create_and_train_model(x_train, x_test, y_train, y_test, len_user, len_courses): 
    model = RecommenderNet(len_user + 1, len_courses + 1, EMBEDDING_SIZE)
    model.compile(
        loss=tf.keras.losses.BinaryCrossentropy(),
        optimizer='adam',
        metrics=[tf.keras.metrics.MeanSquaredError()]
    )

    history = model.fit(
        x=x_train,
        y=y_train,
        batch_size=1024,
        epochs=20,
        verbose=1,
        validation_data=(x_test, y_test),
    )
    return model, history 

def display_prediction(model, x_test, y_test, nums=10):
    pred = model.predict(x_test)
    for i in range(nums): 
        print(f'Predict: {pred[i] * 5.0}')
        print(f'Actual: {y_test[i] * 5.0}')
        print('====================================')
    return pred

# Example usage
len_user, len_courses = get_course_user_len(df)
x_train, x_test, y_train, y_test = get_train_test_data(df)
model, history = create_and_train_model(x_train, x_test, y_train, y_test, len_user, len_courses)
display_prediction(model, x_test, y_test, 20)
```

```python
def get_user_rating_history(user_id, df):
    user_history = df[df['user_id'] == user_id][['course_id', 'name', 'rating']]
    return user_history

def get_course_recommendations(model, user_id, course_ids, top_n=10):
    user_course_pairs = np.array([[user_id, course_id] for course_id in course_ids])
    predictions = model.predict(user_course_pairs).flatten()
    
    top_indices = np.argsort(-predictions)[:top_n]
    recommended_courses = [(course_ids[i], predictions[i] * 5) for i in top_indices]
    return recommended_courses

def recommend_courses_with_description(user_id, course, model, user_input_text, top_n=10):
    user_history = get_user_rating_history(user_id, course)
    print("Riwayat Rating Pengguna:")
    if not user_history.empty:
        print(user_history[['name', 'rating']])
    else:
        print("Pengguna belum memiliki riwayat rating.")
    
    cleaned_input = clean_text(user_input_text)
    user_input_vector = tfidf_vectorizer.transform([cleaned_input])
    similarities = (tfidf_encoded * user_input_vector.T).toarray().flatten()
    
    top_similar_courses = df.iloc[np.argsort(-similarities)[:top_n]]
    similar_course_ids = top_similar_courses['course_id'].values
    
    recommended_courses = get_course_recommendations(model, user_id, similar_course_ids, top_n)
    
    # Filter duplicates based on course_id
    unique_recommendations = []
    seen_course_ids = set()
    
    for course_id, rating_pred in recommended_courses:
        if course_id not in seen_course_ids:
            course_name = course[course['course_id'] == course_id]['name'].values[0]
            unique_recommendations.append((course_name, rating_pred))
            seen_course_ids.add(course_id)
        
        # Stop if we reach the desired number of unique recommendations
        if len(unique_recommendations) >= top_n:
            break

    print("\nRekomendasi Kursus Berdasarkan Deskripsi Masukan:")
    for course_name, rating_pred in unique_recommendations:
        print(f"Kursus: {course_name}, Prediksi Rating: {rating_pred:.2f}")

# Contoh penggunaan
user_id = 1  # Contoh User ID
user_input_text = "data science"
recommend_courses_with_description(user_id, df, model, user_input_text)
```
